```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

$x_i,y_i\in\mathbb{R}^2$  for $i=1,...,n$

```python
n = 1000
X = np.ones([n,2])  # first X[:,0]=1 for the bias
X[:,1] = np.random.rand(n)
Y = np.random.randn(n)

plt.scatter(X[:,1], Y, s=1)
plt.show()
```

# ERM (Empirical Risk Minimization)
$\min_{\beta} \frac{1}{n}\sum_{i=1}^n(\beta_1 x_i+\beta_0-y_i)^2$

Solve normal equation $(X^{\top}X)\beta=X^{\top}Y$

```python
XX = X.transpose().dot(X)
XY = X.transpose().dot(Y)
beta = np.linalg.inv(XX).dot(XY)
print(beta)
```

```python
x_hat = np.linspace(0,1,10)
y_hat = beta[1]*x_hat + beta[0]

plt.scatter(X[:,1], Y, s=1)
plt.plot(x_hat, y_hat, c='r')
plt.show()
```
