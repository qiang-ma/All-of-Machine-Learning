```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# A Simple Version

$x_i\in\mathbb{R}^2$  for $i=1,...,n$

$y_i\in\{-1,1\}$

```python
n = 1500
x = np.ones([2*n, 3])   # add one dimension for the bias
y = np.zeros(2*n)

x[:n,1:] = np.random.normal(2,0.5,[n,2])
y[:n] = np.ones(n)
x[n:,1:] = np.random.normal(0,1,[n,2])
y[n:] = -np.ones(n)

# x0=1 for bias
plt.scatter(x[:n,1],x[:n,2], s=0.5)   # plot all x with y=1
plt.scatter(x[n:,1],x[n:,2], s=0.5)   # plot all x with y=-1
plt.show()
```

## Learning by ERM

To learn $y = \beta^{\top}x$

ERM (Empirical Risk Minimization): $\min_{\beta} \frac{1}{n}\|X\beta-Y\|^2$

Normal Equation: $(X^{\top}X)\beta=X^{\top}Y$

```python
XX = x.transpose().dot(x)
XY = x.transpose().dot(y)
beta = np.linalg.inv(XX).dot(XY)
print(beta)
```

## Optimal Linear Classifier

$y = \begin{cases}
1&\beta^{\top}x\geq0\\
-1&\beta^{\top}x<0
\end{cases}$

```python
def f(x0):
    # beta0 for bias
    x1 = ( - beta[0] - beta[1] * x0) / beta[2]
    return x1
```

```python
plt.scatter(x[:n,1],x[:n,2], s=0.5)   # plot all x with y=1
plt.scatter(x[n:,1],x[n:,2], s=0.5)   # plot all x with y=-1

x_hat_0 = np.linspace(-2, 3.5, 10)
x_hat_1 = f(x_hat_0)
plt.plot(x_hat_0, x_hat_1, c='r')
print("Plot the decision boundary")
plt.show()
```
